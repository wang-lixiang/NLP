```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
# 选择数据集的特征，第4列（花瓣宽度）作为输入特征
x = iris['data'][:,3:]
# 创建二分类目标变量，将目标变量等于2的样本标记为1，其余为0
# .astype(np.int_)：将布尔数组转换为整数数组，其中True转换为1，False转换为0
y = (iris['target'] == 2).astype(np.int_)
x,y
```

```python
# 创建逻辑回归模型，使用sag随机梯度下降，并设置最大迭代次数为1000
binary_classifier = LogisticRegression(solver='sag',max_iter=1000)
# 用特征x和目标变量y训练逻辑回归模型
binary_classifier.fit(x,y)

# 生成从0到3之间的1000个等间距的点，并将其形状调整为列向量
x_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = binary_classifier.predict_proba(x_new)  # 计算新数据点的类别概率
y_hat = binary_classifier.predict(x_new) # 预测新数据点的类别
y_proba,y_hat
```

### 拓展 逻辑回归进行多分类

```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
# 选择数据集的特征，第4列（花瓣宽度）作为输入特征
x = iris['data'][:,3:]
# 会有0，1，2三个类别
y = iris['target']
```

```python
# 创建逻辑回归模型，使用sag随机梯度下降，并设置最大迭代次数为1000 ,one-over-rest
multi_classifier = LogisticRegression(solver='sag',max_iter=1000,multi_class='ovr')
# 用特征x和目标变量y训练逻辑回归模型
multi_classifier.fit(x,y)

# 生成从0到3之间的1000个等间距的点，并将其形状调整为列向量
x_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = multi_classifier.predict_proba(x_new)  # 计算新数据点的类别概率
y_hat = multi_classifier.predict(x_new) # 预测新数据点的类别
y_proba,y_hat
```
