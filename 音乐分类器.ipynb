```python
import numpy as np

# 读取傅里叶变换后的数据集，将其做成机器学习所需要的x,y
genre_list = ["classical", "jazz", "country", "pop", "rock", "metal"]
x = []
y = []
for g in genre_list:
    for n in range(100):
        rad = "./Data/trainset/" + g + "." + str(n).zfill(5) + ".fft.npy"
        # 加载存储在磁盘上的 NumPy 数组文件（.npy 或 .npz 格式）的函数,文件是 .npy 格式，函数将返回一个数组。
        fft_feature = np.load(rad)
        x.append(fft_feature)
        # 返回数值标签
        y.append(genre_list.index(g))
x = np.array(x)
y = np.array(y)
```

```python
from sklearn.linear_model import LogisticRegression
import pickle

# 训练模型并且保存模型，softmax回归，随机梯度下降
model = LogisticRegression(multi_class='multinomial',solver='sag',max_iter=10000)
model.fit(x,y)

output = open('music_class_mode.pkl', 'wb')
pickle.dump(model, output)
output.close()
```

```python
from pprint import pprint

pkl_file = open('music_class_mode.pkl','rb')
# 使用 pickle.load 函数从文件中加载序列化的对象并将其存储在 model_loaded 变量中
model_loaded = pickle.load(pkl_file)
# 使用 pprint 模块的 pprint 函数美观地打印加载的对象，这对于复杂的嵌套结构特别有用
pprint(model_loaded)
pkl_file.close()
```
